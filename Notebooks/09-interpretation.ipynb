## Ground Living area

Ground living area fue transformada logaritmicamente, por lo que cualquier cambio debe ser aplicada la misma transformación. Por ejemplo entonces, si quisieramos ver cuanto afecta un cambio en 100 ft2 en el area habitable al precio, nos daría un incremento en el valor aproximado de 4 USD.

```python
np.expm1((np.log1p(100)/(scaler.var_[1]**0.5))*loaded_model.coef_[0][2])
```
